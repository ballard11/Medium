```python
from pytrends.request import TrendReq
import plotly.express as px

pytrends = TrendReq(hl='en-US', tz=360) 

# build payload
kw_list = ["Data Science"] # list of keywords to get data 
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y') 
```

```python
#1 Interest over Time
data = pytrends.interest_over_time() 
data = data.reset_index() 

fig = px.line(data, x="date", y=['Data Science'], title='Keyword Web Search Interest Over Time')
fig.show() 
```

```python
by_region = pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)
by_region[by_region["Data Science"] > 0]

```
