```python
home_points_list = df['home_points'].tolist()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Assuming home_points_list is your list
pt_count = Counter(home_points_list)
xs = range(101)  # or whatever range is appropriate for your data
ys = [pt_count[x] for x in xs]

plt.bar(xs, ys)
plt.xlabel('Home Points')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Home Points')
plt.show()
```
